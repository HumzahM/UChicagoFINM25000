```python
import numpy as np 
import pandas as pd 
import statistics as stats 
import math
import matplotlib.pyplot as plt

factor_data = pd.read_excel("factor_pricing_data.xlsx", "factors (excess)", usecols="B:E")
factor_data = factor_data.to_numpy()
factor_data = np.transpose(factor_data)
factor_names = ["MKT", "SMB", "HML", "UMD"]
dates = pd.read_excel("factor_pricing_data.xlsx", "factors (excess)", usecols="A", dtype="string")
dates = dates.to_numpy()
dates = np.transpose(dates)

counter = 0
factor_returns = [0, 0, 0, 0]
for column in factor_data:
    mean = stats.mean(column)*12
    factor_returns[counter] = mean
    vol = stats.stdev(column)*math.sqrt(12)
    print(factor_names[counter])
    print("mean: " + str(round(mean,3)))
    print("vol: " + str(round(vol,3)))
    print("SR: " + str(round(mean/vol,3)))
    print("VaR" + str(round(np.quantile(column, 0.05),3)))
    print(" ")
    counter += 1

corr = np.corrcoef(factor_data)
plt.imshow(corr)
plt.xticks(np.arange(4), labels=factor_names)
plt.yticks(np.arange(4), labels=factor_names)
plt.colorbar()

for i in range(4):
    for j in range(4):
        plt.annotate(str(round(corr[i][j], 2)), xy=(i, j),
                     ha='center', va='center', color='white')

counter = 0
plt.figure()
for column in factor_data:
    cumm = np.cumprod(1+column)
    plt.plot(cumm)
    plt.annotate(factor_names[counter], xy=(400, cumm[-1]))
    counter += 1
```

```python
asset_data = pd.read_excel("factor_pricing_data.xlsx", "assets (excess)", usecols="B:M")
asset_data = asset_data.to_numpy()
asset_data = np.transpose(asset_data)
asset_names = ["NoDur", "Durbl", "Manuf", "Enrgy", "Chems", "BusEq", "Telcm", "Utils", "Shops", "Hlth", "Money", "Other"]
asset_returns = np.empty(12)
asset_vol = np.empty(12)
asset_VaR = np.empty(12)
counter = 0
for column in asset_data:
    mean = stats.mean(column)*12
    vol = stats.stdev(column)*math.sqrt(12)
    asset_returns[counter] = mean
    asset_vol[counter] = vol
    VaR = np.quantile(column, 0.05)
    asset_VaR[counter] = VaR
    print(asset_names[counter])
    print("mean: " + str(round(mean,3)))
    print("vol: " + str(round(vol,3)))
    print("SR: " + str(round(mean/vol,3)))
    print("VaR" + str(round(VaR,3)))
    print(" ")
    counter += 1
```

```python
plt.scatter(asset_vol, asset_returns)
plt.xlabel("Asset Vol")
plt.ylabel("Asset Returns")
plt.figure()
plt.scatter(asset_VaR, asset_returns)
plt.xlabel("Asset VaR")
plt.ylabel("Asset Returns")
```

If VaR and Volitality affected the mean return then we would see a trend with increasing volitality and decreasing VaR pointing to more returns. Instead we see that it is scattered which shows that investors care about more than just those two factors.

```python
from sklearn.linear_model import LinearRegression
MAE = np.empty(12)
counter = 0
for column in asset_data:
    model = LinearRegression().fit(factor_data[0].reshape(-1,1),column.reshape(-1,1))
    Beta = model.coef_
    Beta = Beta[0][0]
    Alpha = model.intercept_[0]
    MAE[counter] = abs(Alpha)
    TR = asset_returns[counter]/Beta
    IR = Alpha / asset_vol[counter]
    print(asset_names[counter] + " Alpha: " + str(round(Alpha,4)) + " Beta: " + str(round(Beta,4)) + " TR: " + str(round(TR,4)) + " IR: " + str(round(IR,4)))
    counter += 1
```

If the CAPM was true then the treynor ratio would all be the same (as return is proprtional to vol) and the alphas would be 0

```python
print("MAE: " + str(round(np.mean(MAE)*100,4)) + "%")
```

If the pricing model worked we would expect the MAE to be nearly 0 because there would be no alpha. The MAE is very low in this case and could support the fact that the model works. 

```python
counter = 0
MAE = np.empty(12)
for column in asset_data:
    model = LinearRegression().fit(factor_data.transpose(),column.reshape(-1,1))
    print(asset_names[counter] + " Alpha: " + str(round(model.intercept_[0]*12,5)))
    Alpha = model.intercept_[0]
    MAE[counter] = abs(Alpha)
    counter2 = 0
    for coef in model.coef_[0]:
        print("   " + factor_names[counter2] + " BETA " + str(round(coef,3)))
        counter2 += 1
    print("   R^2 " + str(round(model.score(factor_data.transpose(),column.reshape(-1,1)),4)))
    print("")
    counter += 1
```

```python
print("MAE: " + str(round(np.mean(MAE)*100,4)) + "%")
```

Again, if the pricing model worked we would expect the MAE to be nearly 0 because there would be no alpha. The MAE is very low in this case and could support the fact that the model works.
