```python
#Get Data + Setup 
import numpy as np 
import pandas as pd
import statistics 
import math

data = pd.read_excel('multi_asset_etf_data-1.xlsx', sheet_name='excess returns', usecols="B:L")
data = data.to_numpy()
data = np.transpose(data)
names = ["BWX", "DBC", "EEM", "EFA", "HYG", "IEF", "IYR", "PSP", "QAI", "SPY", "TIP"]
```

```python
counter = 0
returns = np.empty(11)
volatilities = np.empty(11)
for column in data:
    mean_return = 12*statistics.mean(column)
    volitality = math.sqrt(12)*statistics.stdev(column)
    sharpe_ratio = mean_return / volitality
    returns[counter] = mean_return
    volatilities[counter] = volitality 
    print(names[counter] + " Mean Return: " + str(round(mean_return,4)) + " Volitality: " + str(round(volitality,4)) + " Sharpe Ratio: " + str(round(sharpe_ratio, 4)))
    
    counter += 1
```

We can see that SPY and HYG has the best Sharpe Ratio and BWC and DBC have the worst Sharpe Ratios

```python
import matplotlib.pyplot as plt
corr = np.corrcoef(data)
"""
corr2 = np.zeros([11,11])
for i in range(11):
    for j in range(11):
        corr2[j-1][i-1] = corr[i-1][j-1]
"""
plt.imshow(corr)
plt.colorbar()

for i in range(11):
    for j in range(11):
        plt.annotate(str(round(corr[i][j], 2)), xy=(i, j),
                     ha='center', va='center', color='white')
#add annotations 
```

```python
def tangency_portfolio(return_vector, cov_mat):
    inverse_covmat = np.linalg.inv(cov_mat)
    #plt.imshow(inverse_covmat)
    #plt.colorbar()
    #print(inverse_covmat)
    one_vector = (np.ones((11,1), dtype = int))
    sigmaNeg1Mu = inverse_covmat @ return_vector
    scalar = inverse_covmat @ return_vector
    scalar = np.sum(scalar)
    return sigmaNeg1Mu / scalar
    
tangent_weights = tangency_portfolio(returns, corr)

for i in range(11):
    print(names[i] + " Weight: " + str(round(100*tangent_weights[i],1)), "%")
    
#delta is the allocation we choose to get enough mean return in our allocation
#delta is to think about how much risk we get in our allocation 
```

```python
mean = np.dot(np.transpose(returns), tangent_weights)
print("Mean of portfolio: " + str(mean))
volatality = np.dot(np.dot(tangent_weights.transpose(), corr), tangent_weights) ** 0.5
print("Volitality: " + str(volitality))
SR = mean / volitality 
print("Sharpe Ratio: " + str(SR))
```

```python
target = 0.0075*12
scale = target / mean
for i in range(11):
    print(names[i] + " Weight: " + str(round(100*scale*tangent_weights[i],1)), "%")
print("Mean of portfolio: " + str(mean*scale))
print("Volitality: " + str(volitality*scale))
print("Sharpe Ratio: " + str(mean / volitality))
```
